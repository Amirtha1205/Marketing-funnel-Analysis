```python
#task
df['event_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("event type Percentage")
plt.ylabel('') 
plt.show()
```

```python
#observation
1.The view event has the highest percentage at 96.6%, which means most users only view products.
2.The cart event is very low at 1.8%, showing that only a small number of users add products to the cart.
3.The purchase event is the lowest at 1.6%, indicating that very few users complete the buying process.

#Insights
1.There is a high drop-off rate from viewing products to adding them to the cart and purchasing.
2.Users are interested in browsing products, but the conversion rate is low, which may indicate pricing issues, poor user experience, or lack of purchase motivation. 

#business Recommandation

Improve conversion rates by offering:
1.discounts,
2.better product recommendations,
3.simplified checkout process,
4.attractive offers for cart users.
```
